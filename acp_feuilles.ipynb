```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = 'Images 1'
enrg_path = 'img_canny' 
angle=360
```

```python
df=pd.DataFrame(columns=['A','Excentricit√©','Elongation','Surface','Circularity','D','Form Factor','CV'])
```

```python
def moyenne(liste):
    somme=0
    for i in liste:
        somme=somme+i;
    return somme/len(liste)

def distance(point1, point2):
    return np.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)
def distance2(point1,point2):
    return point2-point1
```

```python
if not os.path.exists(enrg_path):
    os.makedirs(enrg_path)  

for i in os.listdir(path):

    img_path = os.path.join(path, i)
    img = cv2.imread(img_path)


    blur = cv2.GaussianBlur(img, (25, 25), 0, sigmaY=0)
    can = cv2.Canny(blur, 50, 150, apertureSize=5, L2gradient=True)
    can2=can.copy()

    base, ext = os.path.splitext(i)  
    enrg = os.path.join(enrg_path, base + "_canny" + ext)  
    print(enrg)
    contours, _ = cv2.findContours(can, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)



   

    max_contour = None
    max_area = 0

    for contour in contours:
        # Calculer l'aire du contour
        area = cv2.contourArea(contour)
        if area > max_area:
            max_area = area
            max_contour = contour

    if max_contour is not None:
        # Calculer le centre du contour
        M = cv2.moments(max_contour)
        if M["m00"] != 0:
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])

            # Trouver le point le plus √©loign√© du centre de gravit√©
            max_distance = 0
            for point in max_contour:
                distance = np.sqrt((point[0][0] - cX)**2 + (point[0][1] - cY)**2)
                if distance > max_distance:
                    max_distance = distance

            # Dessiner un cercle autour du centre de gravit√© avec le rayon correct
            cv2.circle(img, (cX, cY), int(max_distance), (255, 51, 153), 8)  

            # Calculer le rayon du cercle
            radius = int(max_distance)

            # Calculer l'angle entre chaque segment
            angle_step = 360 / 10

            # Dessiner les lignes radiales pour diviser le cercle en 10 parties √©gales
            for angle_deg in range(0, 360, int(angle_step)):
                angle_rad = np.deg2rad(angle_deg)
                x_end = int(cX + radius * np.cos(angle_rad))
                y_end = int(cY + radius * np.sin(angle_rad))
                cv2.line(img, (cX, cY), (x_end, y_end), (255, 51, 153), 8)  



    largest_contour = max(contours, key=cv2.contourArea)
    leaf_area = cv2.contourArea(largest_contour)
    print("Leaf Area (using Contour):", leaf_area)

    points=[]

    indices = np.where(can!= [0])
    coordinates = zip(indices[0], indices[1])
    
    moy_x=moyenne(indices[0])
    moy_y=moyenne(indices[1])

    max_distance = -1
    loin = None

    # radius = distance((moy_x, moy_y), loin)
    w=distance2(min(indices[0]),max(indices[0]))
    l=distance2(min(indices[1]),max(indices[1]))
    exe=w/l

    
    print('exe est ', exe)
    ele=1-exe

    print('ele est', ele)
    circularity=leaf_area/radius**2
    d=np.sqrt(circularity)

    ff=4*3.1415*leaf_area/radius**2

    largest_contour = max(contours, key=cv2.contourArea) # largest contour pour convex hullm, perimeter et circularity
    convex_hull = cv2.convexHull(largest_contour)
    convex_hull_image = cv2.drawContours(img, [convex_hull], -1, (255, 204, 0), 10)
    convex_perimeter = cv2.arcLength(convex_hull, True)
    p = cv2.arcLength(largest_contour, True)
    print('la valeur de p est', p)

    cv = convex_perimeter / p
    print('la valeur de cv est', cv)
    df.loc[i] = [i,exe, ele, leaf_area, circularity, d, ff,cv]

    cv2.circle(can, (int(moy_x), int(moy_y)), int(radius), (255, 204, 0), 10)



    cv2.imwrite(enrg, can)  
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    
    plt.subplot(122),plt.imshow(can2,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])


    plt.show()
    sector_pixel_counts = []
    for i in range(10):
        start_angle = i * 36
        end_angle = (i + 1) * 36

        sector_mask = np.zeros_like(img)
        cv2.ellipse(sector_mask, (cX, cY), (int(radius), int(radius)), 0, start_angle, end_angle, (255, 255, 255), -1)

        sector = cv2.bitwise_and(img, sector_mask)

        pixel_count = np.count_nonzero(sector)
        sector_pixel_counts.append(pixel_count)

        plt.subplot(2, 5, i + 1)
        plt.imshow(sector, cmap='gray')
        plt.title(f'Sector {i + 1}')
        plt.axis('off')
    plt.show()

```

```python
df.set_index('A', inplace=True)
print(df)

data= df.to_numpy()


```

```python
standardized_data = (data - data.mean(axis = 0)) / data.std(axis = 0)
```

```python
### Step 2: Calculate the Covariance Matrix
# use `ddof = 1` if using sample data (default assumption) and use `ddof = 0` if using population data

covariance_matrix = np.corrcoef(standardized_data, ddof = 0, rowvar = False)
print(covariance_matrix)


```

```python
### Step 3: Eigendecomposition on the Covariance Matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)


```

```python
### Step 4: Sort the Principal Components
# np.argsort can only provide lowest to highest; use [::-1] to reverse the list
order_of_importance = np.argsort(eigenvalues)[::-1] 

```

```python
# utilize the sort order to sort eigenvalues and eigenvectors
sorted_eigenvalues = eigenvalues[order_of_importance]
sorted_eigenvectors = eigenvectors[:,order_of_importance] # sort the columns


### Step 5: Calculate the Explained Variance
# use sorted_eigenvalues to ensure the explained variances correspond to the eigenvectors
explained_variance = sorted_eigenvalues / np.sum(sorted_eigenvalues)


### Step 6: Reduce the Data via the Principal Components
k = 2 # select the number of principal components
reduced_data = np.matmul(standardized_data, sorted_eigenvectors[:,:k]) # transform the original data
print(reduced_data)

```

```python
plt.scatter(reduced_data[:,0],reduced_data[:,1])
plt.title('Reduced Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')


plt.show()
```

```python

total_explained_variance = sum(explained_variance[:k])
print(total_explained_variance)
```
